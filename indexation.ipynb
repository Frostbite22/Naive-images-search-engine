```python
a = [[1,1,2,2,2],[0,0,1,1,0],[2,2,3,3,3],[0,1,0,2,3],[3,3,1,2,2]]
```

 ## **Correlogramme d'une image A**

```python
def correlogramme(a):
  correlo = {}
  for i in range(1,len(a[0])-1):
    for j in range(1,len(a[i])-1):
      try:
        correlo[a[i][j]] 
      except:
        correlo[a[i][j]] = {}
      try :
        correlo[a[i][j]][a[i-1][j-1]] += 1
      except:
        correlo[a[i][j]][a[i-1][j-1]] = 1
      try:
        correlo[a[i][j]][a[i-1][j]] += 1
      except:
        correlo[a[i][j]][a[i-1][j]] = 1
      try:
        correlo[a[i][j]][a[i-1][j+1]] += 1
      except:
        correlo[a[i][j]][a[i-1][j+1]] = 1
      try:
        correlo[a[i][j]][a[i][j-1]] += 1
      except:
        correlo[a[i][j]][a[i][j-1]] = 1
      try:
        correlo[a[i][j]][a[i][j+1]] += 1
      except:
        correlo[a[i][j]][a[i][j+1]] = 1
      try:
        correlo[a[i][j]][a[i+1][j-1]] += 1
      except:
        correlo[a[i][j]][a[i+1][j-1]] = 1
      try:
        correlo[a[i][j]][a[i+1][j]] += 1
      except:
        correlo[a[i][j]][a[i+1][j]] = 1
      try:
        correlo[a[i][j]][a[i+1][j+1]] += 1
      except :
        correlo[a[i][j]][a[i+1][j+1]] = 1
  return correlo
```

```python
corrA = correlogramme(a)
```

```python
## on a obtenu les resultats du TD
corrA
```

```python
for i in corrA.keys():
  ax = plt.bar(list(corrA[i].keys()), corrA[i].values(), color='g')
  plt.ylabel('number')
  plt.xlabel('valeur')
  plt.show()
```

## **Descripteur d'une image A**

```python
def descripteur(correlogramme):
  desc = []
  for key in correlogramme:
    for key2 in correlogramme[key]:
      if(key2 == key):
        desc.append(correlogramme[key][key2])
  return desc 
```

```python
desc = descripteur(corrA)
```

```python
## meme resultat que l'example du TD
desc
```

## **Histogramme d'une image A**

```python
def histogramme(a):
  hist = {}
  for i in range(len(a[0])):
    for j in range(len(a[i])):
      try:
        hist[a[i][j]]  += 1
      except:
        hist[a[i][j]]  = 1 
  return hist

```

```python
histA = histogramme(a)
```

```python
plt.bar(list(histA.keys()), histA.values(), color='r')
plt.ylabel('number')
plt.xlabel('valeur')
plt.show()
```

```python
import functools

sommeHistGrayscaleImg1 = functools.reduce(lambda a, b: a+b,histgrayScaleImg.values())
sommeHistGrayscaleImg2 = functools.reduce(lambda a, b: a+b,histgrayScaleImg2.values())
sommeHistGrayscaleImg3 = functools.reduce(lambda a, b: a+b,histgrayScaleImg3.values())
sommeHistGrayscaleImg4 = functools.reduce(lambda a, b: a+b,histgrayScaleImg4.values())
```

```python
sumOfMins = 0
for i in range(0,256):
  try:
    sumOfMins += min(histgrayScaleImg[i],histgrayScaleImg2[i])
  except :
    sumOfMins += 0

dInterIm1_2 = 1 - (sumOfMins/sommeHistGrayscaleImg1)
```

```python
## distance d'intersection image 1 et 2
dInterIm1_2
```

```python
def distance_intersection(histImgReq,histImg):
  sommeHistImgReq = functools.reduce(lambda a, b: a+b,histImgReq.values())
  sumOfMins = 0
  ## puisque j'ai choisit des images codés sur 256 bits c'est évident d'avoir 256 niveaux 
  for i in range(0,256):
    try:
      sumOfMins += min(histImgReq[i],histImg[i])
    except :
      sumOfMins += 0

  dInter = 1 - (sumOfMins/sommeHistImgReq)
  return dInter
```

```python
dInter = []
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg2))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg3))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg4))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg5))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg6))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg7))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg8))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg9))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg10))
dInter.append(distance_intersection(histgrayScaleImg,histgrayScaleImg11))
```
